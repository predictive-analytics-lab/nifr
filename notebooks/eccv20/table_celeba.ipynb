```python
import pandas as pd
```

```python
gender_files = {
#     "cFlow20": "../../experiments/finn/celeba_xy_y_on_recons_rosy-feather-more-eval.csv",
    "cFlow": "../../experiments/finn/celeba_xy_y_on_recons_rosy-feather-epoch60-more-eval.csv",
#     "cFlow140": "../../experiments/finn/celeba_xy_y_on_recons_rosy-feather-epoch140-more-eval.csv",
    "cVAE": "../../experiments/vae/celeba_xy_y_on_recons_gender_more_eval.csv",
    "CNN": "../../experiments/baseline/celeba_naive_baseline_Male_Smiling_40epochs.csv",
    "Kim et al.": "../../experiments/ln2l/celeba_ln2l_Smiling_0.01ew_40epochs_gender_more_eval.csv",
}
```

```python
def read(file_dict):
    df = pd.DataFrame()
    for method, fname in file_dict.items():
        data = pd.read_csv(fname)
        data["method"] = method
        df = pd.concat([df, data], axis="index", sort=False, ignore_index=True)
    return df
```

```python
gender = read(gender_files)
```
