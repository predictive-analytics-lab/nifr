```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("../experiments/finn/xy_recon_aelevels2_huber.csv")
```

```python
fig, plot = plt.subplots(dpi=200, figsize=(3, 2))
plot.plot(data['Scale'], data['Accuracy'], marker='', linestyle="--", linewidth=2.5)
plot.set_xlabel("scale")
plot.set_ylabel("accuracy")
plot.set_title("CMNIST, INN with AE")
plot.grid(True)
```
